# Number of social network users of social media platforms worldwide in 2023 (in millions)

```python
#Here the points are represented individually with a dot, circle, or other shape.
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
#visualization of the data using scattter plot.

x = ["FB","INSTA","TWITTER","SNAPCHAT"]
y = [2023.7,1002.3,309.1,333.4]

plt.plot(x, y,'d',color="purple",alpha=0.3);
plt.xlabel("Social Media")
plt.ylabel("Users in millions");
```
